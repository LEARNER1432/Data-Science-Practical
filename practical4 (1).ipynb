```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('Housing.csv')
df.head(5)

X = df[['area']]
Y = df['price']

train_X, test_X, train_y, test_y = train_test_split(X, Y, train_size=0.8, random_state=100)


df_lm_sklearn = LinearRegression()
df_lm_sklearn.fit(train_X, train_y)

y_pred = df_lm_sklearn.predict(test_X)

mse = mean_squared_error(test_y, y_pred)
r2 = r2_score(test_y, y_pred)

print(f"MSE: {mse}, R-squared: {r2}")

plt.scatter(test_X, test_y, color='blue', label='Actual')
plt.plot(test_X, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.title("Linear Regression Model")
plt.show()
```
