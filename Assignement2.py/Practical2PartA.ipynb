```python
missing_values = ["Na", "na"]
df = pd.read_csv("StudentPerformance.csv", na_values =missing_values)
print(df) #replace na values by NaN
```

```python
print(df.fillna(0)) #fill null values with 0
```

```python
print(df.dropna())
df.dropna(how = 'all')
df.dropna(axis = 1)
```

```python
col=['Math_Score','Reading_Score','Writing_Score','Placement_Score']
df.boxplot(column=col)
plt.show()

# DETECTING OUTLIERS
print(np.where(df['Math_Score']<59))
print(np.where(df['Reading_Score']>95))
print(np.where(df['Writing_Score']<60))
```

```python
# Detecting outliers using Scatterplot
fig, ax = plt.subplots(figsize=(18, 10))
ax.scatter(df['Placement_Score'], df['Placement_offer_Count'])
plt.show()

print(np.where((df['Placement_Score']<50) & (df['Placement_offer_Count']>1)))
print(np.where((df['Placement_Score']>85) & (df['Placement_offer_Count']<3)))
```

```python
# Detecting outliers using Z-Score
from scipy import stats
z = np.abs(stats.zscore(df['Placement_offer_Count']))
print(z)

threshold=1.0
sample_outliers = np.where(z <threshold)
sample_outliers


```

```python
#  Detecting outliers using Inter Quantile Range(IQR)
R = sorted(df['Writing_Score'])
print(R)
q1 = np.percentile(R, 25)
q3 = np.percentile(R, 75)
print(q1,q3)

IQR = q3-q1
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)


r_outliers = []
for i in R:
 if (i<lwr_bound or i>upr_bound):
  r_outliers.append(i)
print(r_outliers)

```

```python
new_df=df
for i in sample_outliers:
 new_df.drop(i,inplace=True)
new_df
```

```python
df_stud=df
ninetieth_percentile = np.percentile(df_stud['Math_Score'], 90)
b = np.where(df_stud['Math_Score']>ninetieth_percentile,
ninetieth_percentile, df_stud['Math_Score'])
print("New array:",b)
df_stud.insert(1,"m score",b,True)
df_stud
```

```python
col = ['Reading_Score']
df.boxplot(col)
median=np.median(R)
median
refined_df=df
refined_df['Reading_Score'] = np.where(refined_df['Reading_Score'] >upr_bound, median,refined_df['Reading_Score'])
print(df)
refined_df['Reading_Score'] = np.where(refined_df['Reading_Score'] <lwr_bound, median,refined_df['Reading_Score'])
print(df)

col = ['Reading_Score']
refined_df.boxplot(col)


```
